### 3. DataFrame Operations

##### - Add a new column named TotalExperience

```python
df["TotalExperience"] = df["Experience"] + df["AdditionalExperience"]
df
```

##### - Modify the 'Salary' of employees with 'Education' as 'WO' to 10% more

```python
df.loc[df["Education"] == "WO", "Salary"] *= 1.10
df
```

#### - Delete rows where 'Role' is 'Microsoft Consultant'

```python
df = df[df["Role"] != "Microsoft Consultant"]
df
```

#### - Create a new column named 'SalaryCategory' with values 'Low', 'Medium', and 'High' based on salary ranges

For this question, you will need to do a self-study on the Cut function from the Pandas library. Discuss it in your group and try to help each other understand its functionality well.

Use these bins and labels for the categories.
- salary_bins = [0, 3000, 5000, 8000]
- salary_labels = ['Low', 'Medium', 'High']

```python
salary_bins = [0, 3000, 5000, 8000]
salary_labels = ['Low', 'Medium', 'High']

df.loc[:, "SalaryCategory"] = pd.cut(df["Salary"], bins=salary_bins, labels=salary_labels)
df
```

#### - Add a new column named 'Bonus' and use NumPy randint to fill it with random integer values

- Generates integers between 100 and 300.

#### NOTE: From now on, your bonus column will have different values than what you see here, because randint generates a new random number each time we run it.
