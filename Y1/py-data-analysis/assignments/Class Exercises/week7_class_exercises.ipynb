```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = iris.data[:, 2:4]  # Selecting "Petal Length" and "Petal Width" features

# K-means clustering function
def kmeans_clustering(k, data):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    return kmeans.labels_, kmeans.cluster_centers_

# Perform K-means clustering
k = 3
assignments, centroids = kmeans_clustering(k, data)

# Visualize the results
plt.scatter(data[:, 0], data[:, 1], c=assignments, cmap='viridis', label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', label='Centroids')
plt.title('K-means Clustering on Iris Dataset (Petal Length vs. Petal Width)')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Load the Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
data = iris.data

# Perform hierarchical clustering
linked = linkage(data, method='ward')

# Plot the dendrogram
plt.figure(figsize=(12, 8))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram for Iris Dataset')
plt.xlabel('Sample Index')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = iris.data

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.7, min_samples=5)
clusters = dbscan.fit_predict(data)

# Visualize the clusters
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis', label='Clusters')
plt.title('DBSCAN Clustering on Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
data = iris.data

# Perform PCA for 2D visualization
pca = PCA(n_components=2)
data_2d = pca.fit_transform(data)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.7, min_samples=5)
clusters = dbscan.fit_predict(data)

# Visualize the clusters
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=clusters, cmap='viridis', label='Clusters')
plt.title('DBSCAN Clustering on Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
