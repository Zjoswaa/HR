## Exercise: Create a Pie Chart

### - Create a pie chart to visualize the distribution of passengers based on the 'Pclass' (Passenger Class) column.

```python
# Count the number of passengers in each class
class_counts = df['Pclass'].value_counts()
class_counts
```

```python
# Create a Pie Chart
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue', 'lightgreen'])
plt.title('Distribution of Passengers by Class')
plt.show()
```

## Exercise: Create a Box Plot

### - Create a box plot to display the distribution of ages ('Age') for different passenger classes ('Pclass').

```python
# plt.boxplot(data)

# plt.boxplot(data1, data2, data3)

data1 = df[df['Pclass'] == 1]['Age']
data2 = df[df['Pclass'] == 2]['Age']
data3 = df[df['Pclass'] == 3]['Age']

plt.boxplot([data1, data2, data3])
plt.show()
```

```python
# Create a Box Plot
plt.boxplot([df[df['Pclass'] == 1]['Age'],
             df[df['Pclass'] == 2]['Age'],
             df[df['Pclass'] == 3]['Age']],
            labels=['1st Class', '2nd Class', '3rd Class'])
plt.title('Age Distribution by Passenger Class')
plt.show()
```

## Exercise: Create a Histogram

### - Create a histogram to show the distribution of fares ('Fare') for the passengers.

```python
# Create a Histogram
plt.hist(df['Fare'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Distribution of Fares')
plt.show()
```

## Exercise: Create a Bar Chart

### - Create a bar chart to compare the number of male and female passengers ('Sex').

```python
# Create a Bar Chart
sex_counts = df['Sex'].value_counts()
plt.bar(sex_counts.index, sex_counts, color=['lightblue', 'lightcoral'])
plt.title('Number of Male and Female Passengers')
plt.show()
```

## Exercise: Create a Scatter Plot

### - Create a scatter plot to explore the relationship between 'Age' and 'Fare'.

```python
# Create a Scatter Plot
plt.scatter(df['Age'], df['Fare'], alpha=0.5, color='green')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.show()
```

### - Create a scatter plot to display the relationship between age and fare for passengers who survived (Survived = 1) and those who did not survive (Survived = 0). Use different colors for each group.

```python
import matplotlib.pyplot as plt

# Filtering data for passengers who survived and those who did not
survived = df[df['Survived'] == 1]
not_survived = df[df['Survived'] == 0]

# Plotting the scatter plot
plt.scatter(not_survived['Age'], not_survived['Fare'], label='Not Survived', alpha=0.7)
plt.scatter(survived['Age'], survived['Fare'], label='Survived', alpha=0.3, marker='s')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare for Survived and Not Survived Passengers')
plt.legend()
plt.show()
```
